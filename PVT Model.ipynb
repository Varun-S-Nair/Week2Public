```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/Users/varun/Desktop/BeaverWorks/week2/PVT data - Sheet2.csv") 
data/=1000
```

```python
data.head(4)
```

```python
data = data[data > 100]
data = data[data < 400]

for key in data.keys():
    data[key].dropna(inplace=True)
    data[key] = data[key].reset_index(drop=True)
data
```

```python
data_avg = data.mean()
data_avg
```

```python
for i in data_avg.keys():
    if data_avg[i]==data_avg.median():
        print(i + ": "+ str(data_avg.median()))
```
